```python
# 将 pic的数据划分为 all resturant drink
import pandas as pd
```

```python
df = pd.read_excel("data/input/pic_feature.xlsx")
```

```python
# filter open data
open_df = df[df['is_open']==1]
print("total pic: {}".format(len(df)),"open pic: {}".format(len(open_df)),sep="\n")
open_df.to_excel('data/yolo11/open_pic_new.xlsx',index=False)
```

```python
# restaurants
res_df = open_df[open_df['categories'].str.contains('restaurant',case=False)]
print("resturant pic:",len(res_df))
res_df.to_excel('data/yolo11/restaurant_new.xlsx',index=False)
```

```python
# drink
temp_df = open_df[~open_df['categories'].str.contains('sushi bar',case=False)]
# print(len(temp_df))
drink_df = temp_df[temp_df['categories'].str.contains('bar|coffee|tea|cafe|pub',case=False)]
# print(len(bar_df))
# bar_df.to_excel('test20250430.xlsx',index=False)
# data verifiy
filter_cate = []
marks = []
for index,row in drink_df.iterrows():
    # print(index,row['categories'])
    category = [i.lstrip().lower() for i in row['categories'].split(",")]
    # bar|coffee|tea|cafe|pub
    line = []
    for drink in ['bar','coffee','tea','cafe','pub']:
        for cate in category:
            if drink in cate:
                if cate not in ["barre classes","bartending schools", "team building activities","barbers",'cheesesteaks','steakhouses','barbeque',
                             'coffee roasteries','cabaret','internet cafes',
                             'bartenders','cafeteria','public services & government',
                             'musical instruments & teachers','public markets','striptease dancers',
                             'bar crawl','amateur sports teams','public art','public transportation',
                             'professional sports teams','hong kong style cafe']:
                    # print(drink,cate)
                    line.append(cate)
    # print(line)
    if len(line) == 0:
        mark = 'no drink'
    else:
        mark = 'drink'
    marks.append(mark)
drink_df['cate_new'] = str(line)
drink_df['mark'] = marks

# # 未过滤的数据
# bar_df.to_excel('test_new20250505.xlsx',index=False)

drink_result = drink_df[drink_df['mark']=='drink']
# 删除cate_new mark变量的数据

drink_result.to_excel('data/yolo11/drink_new.xlsx',index=False)
```
