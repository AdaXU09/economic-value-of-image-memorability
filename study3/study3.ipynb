```python
# 斯皮尔曼计算
spearman_df = pd.DataFrame()
for i in range(33,49):
    spearman_df['yinxiang_{}'.format(str(i))] = naodao.iloc[:,i+2].values.tolist()+niming.iloc[:,i].values.tolist()

pic_true = [True,False,False,False,False,False,False,False,
       True,True,True,True,True,False,True,True]
print(len(spearman_df))
result = []
for index,row in spearman_df.iterrows():
    # print(index+1,'-------------------------------')
    line = list(row.values)
    # print(line)
    for i in range(len(pic_true)):
        if line[i]=='见过':
            mark = 1
        else:
            mark = 0
        # print([index+1,i+1,pic_true[i],mark])
        result.append([index+1,i+1,pic_true[i],mark])
data = pd.DataFrame(result,columns=['participant_id','image_id','is_target','response'])

# Calculate hit rate (HR) and false alarm rate (FA) for each image
hit_rate = data[data['is_target'] == True].groupby('image_id')['response'].mean().reset_index(name='HR')
#print(hit_rate)
false_alarm_rate = data[data['is_target'] == False].groupby('image_id')['response'].mean().reset_index(name='FA')

# print(false_alarm_rate)
# Merge HR and FA into a single DataFrame
image_scores = pd.merge(hit_rate, false_alarm_rate, on='image_id', how='outer').fillna(0)
image_scores = image_scores.sort_values('image_id')
print(image_scores)
# Calculate CR Score for each image
image_scores['CR_Score'] = round(image_scores['HR'] - image_scores['FA'],3)
memory_score = [0.951,0.617,0.872,0.946,
         0.671,0.965,0.603,0.749,
         0.62,0.978,0.772,0.94,
         0.847,0.772,0.946,0.692]
CR_Score = image_scores['CR_Score'].values.tolist()
print(image_scores['CR_Score'].values.tolist())
print(memory_score)

# HR_target2 = data.loc[data['is_target'] == True].groupby('image_id')['response'].mean().sort_index()
# FA_overall2 = data.loc[data['is_target'] == False, 'response'].mean()

# CR_target2 = (HR_target2 - FA_overall2).round(3)   # length = 8 targets
# print(CR_target2)

from scipy.stats import spearmanr

# 使用spearmanr()函数计算斯皮尔曼等级相关系数
correlation_coefficient, p_value = spearmanr(memory_score, CR_Score)

print("Spearman's correlation coefficient:", correlation_coefficient)
print("p-value:", p_value)


from scipy.stats import spearmanr

a = image_scores['CR_Score'].values.tolist()
b = memory_score
# c = CR_target2
a_filter = []
b_filter = []
for i in range(len(a)):
    if a[i]>0:
        a_filter.append(a[i])
        b_filter.append(b[i])

# 使用spearmanr()函数计算斯皮尔曼等级相关系数
correlation_coefficient, p_value = spearmanr(a_filter, b_filter)

print("Spearman's correlation coefficient:", correlation_coefficient)
print("p-value:", p_value)

# correlation_coefficient, p_value = spearmanr(CR_target2, b_filter)

# print("Spearman's correlation coefficient:", correlation_coefficient)
# print("p-value:", p_value)
# ====================== 主要增强：添加置信区间与统计报告 ======================

def bootstrap_spearman_ci(x, y, alpha=0.05, n_bootstraps=10000):
    """
    使用自助法（bootstrap）计算 Spearman 相关系数的置信区间
    """
    corrs = []
    for _ in range(n_bootstraps):
        idx = np.random.choice(len(x), size=len(x), replace=True)
        x_sample = [x[i] for i in idx]
        y_sample = [y[i] for i in idx]
        rho, _ = spearmanr(x_sample, y_sample)
        corrs.append(rho)
    lower_percentile = 100 * (alpha / 2)
    upper_percentile = 100 * (1 - alpha / 2)
    ci_lower = np.percentile(corrs, lower_percentile)
    ci_upper = np.percentile(corrs, upper_percentile)
    return ci_lower, ci_upper, corrs

# 过滤掉 CR_Score <= 0 的点（如您所做的）
# a = CR_Score
# b = memory_score
# a_filter = []
# b_filter = []
# for i in range(len(a)):
#     if a[i] > 0:
#         a_filter.append(a[i])
#         b_filter.append(b[i])

# # 计算原始 Spearman 相关性
# rho, p_val = spearmanr(a_filter, b_filter)
# n_filtered = len(a_filter)

# np.random.seed(66)
# # 计算置信区间
# ci_lower, ci_upper, boot_corrs = bootstrap_spearman_ci(a_filter, b_filter, alpha=0.05)

a = CR_Score
b = memory_score
a_filter = []
b_filter = []
for i in range(len(a)):
    if a[i] > 0:
        a_filter.append(a[i])
        b_filter.append(b[i])

# 计算原始 Spearman 相关性
rho, p_val = spearmanr(a_filter, b_filter)
n_filtered = len(a_filter)

np.random.seed(66)
# 计算置信区间
ci_lower, ci_upper, boot_corrs = bootstrap_spearman_ci(a_filter, b_filter, alpha=0.05)

# ====================== 输出完整结果（满足审稿人要求）======================

print("\n" + "="*60)
print("             Spearman Correlation Report")
print("="*60)
print(f"Original N (images): {len(memory_score)}")
print(f"Filtered N (CR_Score > 0): {n_filtered}")
print(f"Spearman's ρ: {rho:.3f}")
print(f"95% Confidence Interval: [{ci_lower:.3f}, {ci_upper:.3f}]")
print(f"P-value: {p_val:.4f}")

# (b) 警告小样本可能高估相关性
if n_filtered <= 20:
    print("注意：样本量较小（N = {}），相关系数可能不稳定且被高估。".format(n_filtered))
    print("   建议在正文中注明此局限性，并谨慎解释高相关性结果。")

# (c) 添加文献对比（供您写论文时参考）
print("文献对比参考（用于论文写作）:")
print("Needell & Bainbridge (ResMem): ρ typically ~0.4–0.7")
print("Davis & Bainbridge (art memory): similar mid-range correlations")

# 可选：绘制直方图查看 bootstrap 分布（需 matplotlib）
# import matplotlib.pyplot as plt
# plt.hist(boot_corrs, bins=50, alpha=0.7, color='skyblue')
# plt.axvline(rho, color='red', linestyle='--', label=f'Observed ρ = {rho:.3f}')
# plt.axvline(ci_lower, color='gray', linestyle=':', label='95% CI')
# plt.axvline(ci_upper, color='gray', linestyle=':')
# plt.title("Bootstrap Distribution of Spearman's ρ")
# plt.xlabel("Spearman Correlation Coefficient")
# plt.ylabel("Frequency")
# plt.legend()
# plt.show()
```

```python
# OLS per category — harmonized with LME (Table 5) reference categories
# ref: memorability=High, quality=Low, mood=Very negative
HIGH_SCORES = {'wine': 0.946, 'cake': 0.951, 'sandwich': 0.978, 'donut': 0.94}

def models_mood_detail(data, mark):
    df1 = data.copy()
    # 1=low, 0=high → matches LME ref=high
    df1['low_memo'] = (df1['{}_score'.format(mark)] != HIGH_SCORES[mark]).astype(int)
    for col in ["negative", "neutral", "positive", "very_positive",
                "{}_median".format(mark), "{}_high".format(mark),
                "{}_attr".format(mark)]:
        df1[col] = df1[col].replace({True: 1, False: 0})
    df1 = df1.rename(columns={mark: 'y',
                              '{}_median'.format(mark): 'median_q',
                              '{}_high'.format(mark): 'high_q',
                              '{}_attr'.format(mark): 'attr'})
    model_formula = 'y ~ low_memo + negative + neutral + positive + very_positive + median_q + high_q + attr'
    model = smf.ols(formula=model_formula, data=df1)
    results = model.fit()
    print(results.summary())

for i in ['wine','cake','sandwich','donut']:
    print("********************************************{}******************************************************************".format(i))
    models_mood_detail(final_res, i)
```

```python
# generate latex — harmonized with LME reference categories
# Reference levels (matching model.Rmd):
#   Memorability: High (ref) → reports Low memorability dummy (binary)
#   Quality: Low quality (ref) → reports Medium quality, High quality dummies
#   Mood: Very negative (ref) → reports Negative, Neutral, Positive, Very positive dummies
#   Attractiveness: binary (1 = image rated more attractive than paired counterpart)

HIGH_SCORES = {'wine': 0.946, 'cake': 0.951, 'sandwich': 0.978, 'donut': 0.94}

def models_mood_harmonized(data, mark):
    df1 = data.copy()

    # Create binary memorability dummy (1=low, 0=high) — matches LME ref=high
    df1['low_memo'] = (df1['{}_score'.format(mark)] != HIGH_SCORES[mark]).astype(int)

    # Convert boolean to int
    for col in ["negative", "neutral", "positive", "very_positive",
                "{}_median".format(mark), "{}_high".format(mark),
                "{}_attr".format(mark)]:
        df1[col] = df1[col].replace({True: 1, False: 0})

    df1 = df1.rename(columns={mark: 'y',
                              '{}_median'.format(mark): 'median_q',
                              '{}_high'.format(mark): 'high_q',
                              '{}_attr'.format(mark): 'attr'})

    # ref: memorability=high, quality=low, mood=very_negative
    model_formula = 'y ~ low_memo + negative + neutral + positive + very_positive + median_q + high_q + attr'
    model = smf.ols(formula=model_formula, data=df1)
    results = model.fit()
    return results

wine = models_mood_harmonized(final_res, 'wine')
cake = models_mood_harmonized(final_res, 'cake')
sandwich = models_mood_harmonized(final_res, 'sandwich')
donut = models_mood_harmonized(final_res, 'donut')

# --- Custom LaTeX with correct p-value stars ---

def _stars(p):
    if p < 0.001: return '***'
    if p < 0.01:  return '**'
    if p < 0.05:  return '*'
    if p < 0.10:  return '$^\\dagger$'
    return ''

def _fmt(val, decimals=4):
    return f"{val:.{decimals}f}"

models = [wine, cake, sandwich, donut]
col_names = ['Wine', 'Cake', 'Sandwich', 'Donut']
var_order = list(models[0].params.index)
BS = chr(92) + chr(92)  # LaTeX row ending

rows = []
for var in var_order:
    coef_cells, se_cells = [], []
    for m in models:
        if var in m.params.index:
            coef_cells.append(f"{_fmt(m.params[var])}{_stars(m.pvalues[var])}")
            se_cells.append(f"({_fmt(m.bse[var])})")
        else:
            coef_cells.append('')
            se_cells.append('')
    rows.append((var, coef_cells, se_cells))

lines = []
lines.append(r'\begin{table}[htbp]')
lines.append(r'\centering')
lines.append(r'\caption{OLS Regression Results: Willingness to Pay (Study 3)}')
lines.append(r'\label{tab:study3_ols}')
lines.append(r'\small')
lines.append(r'\begin{tabular}{l' + ' c' * len(models) + '}')
lines.append(r'\toprule')
lines.append(' & '.join([''] + col_names) + ' ' + BS)
lines.append(r'\midrule')

for var, coef_cells, se_cells in rows:
    var_tex = var.replace('_', r'\_')
    lines.append(' & '.join([var_tex] + coef_cells) + ' ' + BS)
    lines.append(' & '.join([''] + se_cells) + ' ' + BS)

lines.append(r'\midrule')
n_cells = [f"{int(m.nobs):,}" for m in models]
r2_cells = [_fmt(m.rsquared, 3) for m in models]
r2_adj_cells = [_fmt(m.rsquared_adj, 3) for m in models]
lines.append(' & '.join([r'$N$'] + n_cells) + ' ' + BS)
lines.append(' & '.join([r'$R^2$'] + r2_cells) + ' ' + BS)
lines.append(' & '.join([r'Adj. $R^2$'] + r2_adj_cells) + ' ' + BS)
lines.append(r'\bottomrule')
lines.append(r'\end{tabular}')
lines.append(r'\begin{minipage}{\textwidth}')
lines.append(r'\vspace{4pt}')
lines.append(r'\footnotesize')
lines.append(r'\textit{Note.} Standard errors in parentheses. '
             r'Reference categories: memorability = High; quality = Low; mood = Very negative. '
             r"``attr'' (more attractive) is a dummy = 1 if the image is rated more attractive "
             r'than its paired counterpart. '
             r'$^\dagger p < .10$, $^* p < .05$, $^{**} p < .01$, $^{***} p < .001$.')
lines.append(r'\end{minipage}')
lines.append(r'\end{table}')

table_tex = '\n'.join(lines)
print(table_tex)

with open('data/output/study3.tex', 'w', encoding='utf-8') as f:
    f.write(table_tex)
print("\nSaved: data/output/study3.tex")
```
