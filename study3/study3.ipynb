```python
# 斯皮尔曼计算
spearman_df = pd.DataFrame()
for i in range(33,49):
    spearman_df['yinxiang_{}'.format(str(i))] = naodao.iloc[:,i+2].values.tolist()+niming.iloc[:,i].values.tolist()

pic_true = [True,False,False,False,False,False,False,False,
       True,True,True,True,True,False,True,True]
print(len(spearman_df))
result = []
for index,row in spearman_df.iterrows():
    # print(index+1,'-------------------------------')
    line = list(row.values)
    # print(line)
    for i in range(len(pic_true)):
        if line[i]=='见过':
            mark = 1
        else:
            mark = 0
        # print([index+1,i+1,pic_true[i],mark])
        result.append([index+1,i+1,pic_true[i],mark])
data = pd.DataFrame(result,columns=['participant_id','image_id','is_target','response'])

# Calculate hit rate (HR) and false alarm rate (FA) for each image
hit_rate = data[data['is_target'] == True].groupby('image_id')['response'].mean().reset_index(name='HR')
# print(hit_rate)
false_alarm_rate = data[data['is_target'] == False].groupby('image_id')['response'].mean().reset_index(name='FA')
# print(false_alarm_rate)
# Merge HR and FA into a single DataFrame
image_scores = pd.merge(hit_rate, false_alarm_rate, on='image_id', how='outer').fillna(0)
image_scores = image_scores.sort_values('image_id')
print(image_scores)
# Calculate CR Score for each image
image_scores['CR_Score'] = round(image_scores['HR'] - image_scores['FA'],3)
memory_score = [0.951,0.617,0.872,0.946,
         0.671,0.965,0.603,0.749,
         0.62,0.978,0.772,0.94,
         0.847,0.772,0.946,0.692]
CR_Score = image_scores['CR_Score'].values.tolist()
print(image_scores['CR_Score'].values.tolist())
print(memory_score)

HR_target2 = data.loc[data['is_target'] == True].groupby('image_id')['response'].mean().sort_index()
FA_overall2 = data.loc[data['is_target'] == False, 'response'].mean()

CR_target2 = (HR_target2 - FA_overall2).round(3)   # length = 8 targets

from scipy.stats import spearmanr

# 使用spearmanr()函数计算斯皮尔曼等级相关系数
correlation_coefficient, p_value = spearmanr(memory_score, CR_Score)

print("Spearman's correlation coefficient:", correlation_coefficient)
print("p-value:", p_value)


from scipy.stats import spearmanr

a = image_scores['CR_Score'].values.tolist()
b = memory_score
a_filter = []
b_filter = []
for i in range(len(a)):
    if a[i]>0:
        a_filter.append(a[i])
        b_filter.append(b[i])

# 使用spearmanr()函数计算斯皮尔曼等级相关系数
correlation_coefficient, p_value = spearmanr(a_filter, b_filter)

print("Spearman's correlation coefficient:", correlation_coefficient)
print("p-value:", p_value)

correlation_coefficient, p_value = spearmanr(CR_target2, b_filter)

print("Spearman's correlation coefficient:", correlation_coefficient)
print("p-value:", p_value)

# ====================== 主要增强：添加置信区间与统计报告 ======================

def bootstrap_spearman_ci(x, y, alpha=0.05, n_bootstraps=10000):
    """
    使用自助法（bootstrap）计算 Spearman 相关系数的置信区间
    """
    corrs = []
    for _ in range(n_bootstraps):
        idx = np.random.choice(len(x), size=len(x), replace=True)
        x_sample = [x[i] for i in idx]
        y_sample = [y[i] for i in idx]
        rho, _ = spearmanr(x_sample, y_sample)
        corrs.append(rho)
    lower_percentile = 100 * (alpha / 2)
    upper_percentile = 100 * (1 - alpha / 2)
    ci_lower = np.percentile(corrs, lower_percentile)
    ci_upper = np.percentile(corrs, upper_percentile)
    return ci_lower, ci_upper, corrs

# 过滤掉 CR_Score <= 0 的点（如您所做的）
a = CR_Score
b = memory_score
a_filter = []
b_filter = []
for i in range(len(a)):
    if a[i] > 0:
        a_filter.append(a[i])
        b_filter.append(b[i])

# 计算原始 Spearman 相关性
rho, p_val = spearmanr(a_filter, b_filter)
n_filtered = len(a_filter)

rho_2, p_val_2 = spearmanr(np.asarray(CR_target2), b_filter)
n_filtered_2 = len(np.asarray(CR_target2))

np.random.seed(66)
# 计算置信区间
# ci_lower, ci_upper, boot_corrs = bootstrap_spearman_ci(a_filter, b_filter, alpha=0.05)
ci_lower_2, ci_upper_2, boot_corrs_2 = bootstrap_spearman_ci(np.asarray(CR_target2, dtype=float), np.asarray(b_filter, dtype=float), alpha=0.05)
print(np.asarray(CR_target2))
print(b_filter)
# ====================== 输出完整结果（满足审稿人要求）======================

print("\n" + "="*60)
print("             Spearman Correlation Report")
print("="*60)
print(f"Original N (images): {len(memory_score)}")
print(f"Filtered N (CR_Score > 0): {n_filtered}")
print(f"Spearman's ρ: {rho_2:.3f}")
print(f"95% Confidence Interval: [{ci_lower_2:.3f}, {ci_upper_2:.3f}]")
print(f"P-value: {p_val_2:.4f}")

# (b) 警告小样本可能高估相关性
if n_filtered <= 20:
    print("注意：样本量较小（N = {}），相关系数可能不稳定且被高估。".format(n_filtered))
    print("   建议在正文中注明此局限性，并谨慎解释高相关性结果。")

# (c) 添加文献对比（供您写论文时参考）
print("文献对比参考（用于论文写作）:")
print("Needell & Bainbridge (ResMem): ρ typically ~0.4–0.7")
print("Davis & Bainbridge (art memory): similar mid-range correlations")

# 可选：绘制直方图查看 bootstrap 分布（需 matplotlib）
# import matplotlib.pyplot as plt
# plt.hist(boot_corrs, bins=50, alpha=0.7, color='skyblue')
# plt.axvline(rho, color='red', linestyle='--', label=f'Observed ρ = {rho:.3f}')
# plt.axvline(ci_lower, color='gray', linestyle=':', label='95% CI')
# plt.axvline(ci_upper, color='gray', linestyle=':')
# plt.title("Bootstrap Distribution of Spearman's ρ")
# plt.xlabel("Spearman Correlation Coefficient")
# plt.ylabel("Frequency")
# plt.legend()
# plt.show()
```

```python
# generate latex
def models_mood(data,mark):
    df1 = data
    df1[["positive","very_positive","negative","very_negative","{}_low".format(mark),"{}_high".format(mark),"{}_attr".format(mark)]] = df1[["positive","very_positive","negative","very_negative","{}_low".format(mark),"{}_high".format(mark),"{}_attr".format(mark)]].replace({True:1,False:0})
    df1 = df1.rename(columns={mark: 'y',
                              '{}_score'.format(mark): 'score',
                              '{}_low'.format(mark): 'low',
                              '{}_high'.format(mark): 'high',
                              '{}_attr'.format(mark): 'attr',
                            #   '{}_pic2'.format(mark): 'pic2',
                              })
    model_formula = 'y~score+positive+very_positive+negative+very_negative+low+high+attr'.format(mark,mark,mark)
    model = smf.ols(formula=model_formula, data=df1)
    results = model.fit()
    return results

wine = models_mood(final_res,'wine')
cake = models_mood(final_res,'cake')
sandwich = models_mood(final_res,'sandwich')
donut = models_mood(final_res,'donut')

# generate LaTeX table
results_table = summary_col([wine, cake,sandwich,donut], stars=True, model_names=['wine', 'cake','sandwich','donut'])
print(results_table)
latex_table = results_table.as_latex()
# latex_table
# # save .tex file
with open('data/output/study3.tex', 'w') as f:
    f.write(latex_table)
```
